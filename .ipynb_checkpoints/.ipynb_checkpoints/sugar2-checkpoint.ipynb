```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

print("Все библиотеки успешно импортированы!")
```

```python
df = pd.read_csv(r"C:\Users\Мартинсон Диана\Downloads\diabetes.csv")
# Покажем первые строки данных
print(df.head())
print(df.shape)
print(df.info())
print(df.describe())
```

```python
columns_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[columns_with_zero] = df[columns_with_zero].replace(0, np.nan)

print("Нулевые значения заменены на NaN!")

print(df.isnull().sum)
```

```python
for column in columns_with_zero:
    df[column].fillna(df.groupby('Outcome')[column].transform('median'))
print(df.describe)
```

```python
no_diabetes = df[df['Outcome'] == 0]
yes_diabetes = df[df['Outcome'] == 1]

print(f"Количество здоровых: {len(no_diabetes)}")
print(f"Количество больных диабетом: {len(yes_diabetes)}")
```

```python
comparison_df = pd.DataFrame({'Здоровые': no_diabetes.mean(), 'Больные диабетом': yes_diabetes.mean()})
print(comparison_df[['Здоровые', 'Больные диабетом']].round(2))
```

```python
import os
import glob

# Поиск ВСЕХ .ipynb файлов в текущей папке и подпапках
print("🔍 Поиск всех .ipynb файлов...")
all_ipynb_files = glob.glob("**/*.ipynb", recursive=True)

for file in all_ipynb_files:
    if "checkpoint" not in file:
        print(f"✅ ОСНОВНОЙ ФАЙЛ: {file}")
    else:
        print(f"📋 checkpoint: {file}")

if not all_ipynb_files:
    print("❌ Файлы .ipynb не найдены!") 
```

```python
import os
import shutil

# Восстановление основного файла из checkpoint
checkpoint_path = "sugar2-checkpoint.ipynb"
main_file_path = "sugar2.ipynb"

if os.path.exists(checkpoint_path):
    # Копируем checkpoint обратно в основной файл
    shutil.copy2(checkpoint_path, main_file_path)
    print(f"✅ ВОССТАНОВЛЕНО: {checkpoint_path} → {main_file_path}")
else:
    print("❌ Checkpoint файл не найден!")
```
